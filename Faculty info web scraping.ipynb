# <b> <center> TRAVERSAL </b> </center>

```python
response = requests.get(url)
response.status_code
```

```python
print(response.text)
```

```python
# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

#print(soup)
print(soup)
```

```python
# Extracting all the links with text Faculty
```

```python
# Filter and print links containing the keyword "faculty"

# Find all anchor (<a>) elements
anchor_elements = soup.find_all('a')
keyword = "faculty"
faculty_links = []

for anchor in anchor_elements:
    href = anchor.get('href')
    if href and keyword in href.lower():
            faculty_links.append(href)

faculty_links
```

```python
for i in faculty_links:
    print(url+i)
```

```python
################### GENERAL CODE

# Find and extract faculty information
faculty_list = []
faculty_elements = soup.find_all('div', class_='faculty-entry')  # Have to adjust the class
```

```python
## For particular faculty entry class , will extract name and email
NAME=[]
EMAIL=[]
for faculty in faculty_elements:
    faculty_name = faculty.find('h2').text.strip()
    faculty_email_element = faculty.find('a', class_='email')  # HAVE TO ADJUST THE CLASS
    faculty_email = faculty_email_element['href'].split(':')[1] if faculty_email_element else "N/A"
    
    faculty_info = {'Name': faculty_name,'Email': faculty_email}

    faculty_list.append(faculty_info)
    NAME.append(faculty_name)
    EMAIL.append(faculty_email)

    # Print the faculty list
for faculty_info in faculty_list:
    print(f"Name: {faculty_info['Name']}")
    print(f"Email: {faculty_info['Email']}")

    

    
    
##Creating the final df
df = pd.DataFrame({'Name': NAME},{'Email': EMAIL})

# Save the DataFrame to a CSV file
df.to_csv('Faculty_info.csv', index=False)
```
